# Task 01: Visualizing Age Distribution in India's Population (2022)
This task involves creating a bar chart to visualize the distribution of India's population by age group, using data from 2022.

**Age Groups and Population Data:**
- 0 to 20 Years: 512 million (36.1%)
- 21 to 64 Years: 807 million (57.0%)
- 65+ Years: 98 million (6.9%)


```python
import matplotlib.pyplot as plt

# Age group data
age_groups = ['0 to 20 Years', '21 to 64 Years', '65+ Years']
population_mn = [512, 807, 98]
percentages = [36.1, 57.0, 6.9]

# Create bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(age_groups, population_mn, color=['gold', 'deepskyblue', 'hotpink'])

# Add labels
for bar, percent in zip(bars, percentages):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{percent}%', ha='center', va='bottom', fontsize=12)

plt.title("India's Population Distribution by Age in 2022")
plt.xlabel("Age Group")
plt.ylabel("Population (in millions)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Data
age_groups = ['0 to 20 Years', '21 to 64 Years', '65+ Years']
population_mn = [512, 807, 98]
percentages = [36.1, 57.0, 6.9]

# Slicer widget
age_selector = widgets.SelectMultiple(
    options=age_groups,
    value=tuple(age_groups),
    description='Select Ages:',
    disabled=False
)

def plot_filtered_chart(selected_ages):
    clear_output(wait=True)
    display(age_selector)
    indices = [age_groups.index(age) for age in selected_ages]
    filtered_groups = [age_groups[i] for i in indices]
    filtered_pop = [population_mn[i] for i in indices]
    filtered_percent = [percentages[i] for i in indices]

    plt.figure(figsize=(8, 6))
    bars = plt.bar(filtered_groups, filtered_pop, color=['gold', 'deepskyblue', 'hotpink'][:len(filtered_groups)])
    for bar, percent in zip(bars, filtered_percent):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, height, f'{percent}%', ha='center', va='bottom', fontsize=12)
    plt.title("Filtered Population Distribution by Age - India 2022")
    plt.xlabel("Age Group")
    plt.ylabel("Population (in millions)")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

age_selector.observe(lambda change: plot_filtered_chart(age_selector.value), names='value')
plot_filtered_chart(age_selector.value)
```
